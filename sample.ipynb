```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
rng = np.random.default_rng(3)
N = 50
df = pd.DataFrame({
    'X': np.arange(N) + 1, 
    'Gamma': rng.gamma(1, size=N),
    'Beta': rng.binomial(8, 0.2, size=N),
    'Uniform': rng.uniform(0, 1, size=N)
})
df.plot(x='X')
```
