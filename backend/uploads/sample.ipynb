```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
rng = np.random.default_rng(3)
N = 100
df = pd.DataFrame({
    'X': np.arange(N) + 1, 
    'Gamma': rng.gamma(1, size=N) + 3,
    'Beta': rng.binomial(9, 0.2, size=N)
})
df.plot(kind='line', x='X', figsize=(8, 4))
```

```python
longer = (
    pd.melt(df, 
            id_vars=['X'], 
            value_vars=['Gamma', 'Beta'],
            var_name='Distribution')
    )
df.set_index('X').plot(kind='density')
```

```python
rng = np.random.default_rng(3)
N = 100
df = pd.DataFrame({
    'X': np.arange(N) + 1, 
    'Gamma': rng.gamma(1, size=N) + 50,
    'Beta': rng.binomial(9, 0.2, size=N),
    'Uniform': rng.uniform(-8, 2, size=N)
})
df.plot(kind='line', x='X', figsize=(8, 4))
```
