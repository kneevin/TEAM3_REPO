```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
rng = np.random.default_rng(3)
N = 100
df = pd.DataFrame({
    'X': np.arange(N) + 1, 
    'Gamma': rng.gamma(1, size=N) + 2,
    'Beta': rng.binomial(8, 0.2, size=N),
    'Uniform': rng.uniform(-8, 2, size=N)
})
df['Combined'] = (df['Gamma'] + df['Beta'] + df['Uniform']) / 3
df.plot(kind='line', x='X', figsize=(5, 10))
```

```python
rng = np.random.default_rng(3)
N = 100
df = pd.DataFrame({
    'X': np.arange(N) + 1, 
    'Gamma': rng.gamma(1, size=N) + 2,
    'Beta': rng.binomial(8, 0.2, size=N),
    'Uniform': rng.uniform(-8, 2, size=N)
})
df['Average'] = (df['Gamma'] + df['Beta'] + df['Uniform']) / 3
df.plot(kind='line', x='X', figsize=(8, 4))
```
